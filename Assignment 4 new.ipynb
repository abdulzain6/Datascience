```python
import pandas as pd
```

```python
df = pd.read_csv("netflix_data.csv")
df
```

```python
def split_comma(data):
  data = str(data)
  if data == "nan":
    return 0
  return len(data.split(","))
```

```python
df["num_of_cast"] = df["cast"].apply(split_comma)
```
